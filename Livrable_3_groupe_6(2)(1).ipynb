## <font color="pink">Calculer la vitesse finale (vf)​ de la voiture à la fin de la pente, sans frottement : </font>

$\sum{Fext}=m\vec{a}$

$a$ (acceleration de la voiture) = $5.8  m/s$

$s$ (longueur de pente) = $31$ m

On calcule d'apport vitesse due à la gravité :

$V_g = 2 . g . h$ 

On calcule vitesse due à l'accélération moyenne : 

$V_a = 2 . a . s$

La vitesse finale en fin de pente est la somme des deux vitesses : 

$V_f = \sqrt{V_a + V_g}$

## <font color="pink"> Calcule vitesse (Vmin) pour entrée le looping </font> : 


- En utilisant la formule de conservation de l'énergie mécanique :

La hauteur (h) du sommet du looping par rapport à son point d'entrée est (2r), où (r) est le rayon du looping. 

En utilisant la formule de conservation de l'énergie mécanique :

$\frac{1}{2}m v_h^2 - \frac{1}{2}m v_m^2= -mg2r + ma . \pi r$ 

Où :
- $m$ est la masse de la voiture (qui se simplifie de chaque côté de l'équation),
- $g$ est l'accélération due à la gravité $9.81 \, \text{m/s}^2$,
- $h = 2r$ est la hauteur du sommet du looping par rapport au point d'entrée.
- $v_h$ est la vitesse minimale au sommet de looping.

En réarrangeant pour trouver $v_{entrée, min}$ :

$v_{min} = \sqrt{v_{h}^2 + 4gr - 2a\pi r}$

En utilisant $v_{sommet} = 7.67 \, \text{m/s}$ et $h = 2r$, calculons $v_{min}$ en supposant un rayon (r) typique pour les modèles de voiture dans le looping. Si $r = 6 \,{m}$, alors $h = 12 \,{m}$. Calculons la vitesse minimale d'entrée requise.

## <font color="pink">Calcule Vf sortie de looping sans frottement : <font> 

En utilisant la formule de conservation de l'énergie mécanique :

$\frac{1}{2}mV_i^2 + mgh_i = \frac{1}{2}mV_f^2 + mgh_f$
    
$\frac{1}{2}mV_i^2 + mgh_i - mgh_f = \frac{1}{2}mV_f^2$    
    
$V_f = \sqrt{V_i^2 + 2g(h_i - h_f)}$
    
$V_f = \sqrt{V_i^2 + 2gh}$

Où :
- $V_f$ est la vitesse de sortie.
- $V_i$ est la vitesse initiale $19,97 m/s$.
- $g$ est l'accélération due à la gravité $9,81 m/s²$.
- $h$ est la différence de hauteur entre les points d'entrée et de sortie du looping.


En utilisant cette formule, avec une vitesse initiale $V_i$ de $19,97 m/s$, la vitesse de sortie $V_f$ après avoir terminé le looping est d'environ $25,18 m/s$.

```python
import math

# Constants
Vi = 19.97  # initial velocity in m/s
g = 9.81    # acceleration due to gravity in m/s^2
h=12

# Recalculate the final velocity Vf
V_f = math.sqrt(Vi**2 + 2 * g * h)

# Display the new final velocity
print("la vitesse de sortie finale apres aprés avoir términer le looping est d'environ : ", V_f)
```

# <font color="pink"> Le Temp total sans frottement  : </font>

Pour calculer le temps nécessaire pour chaque étape de la course, nous pouvons utiliser la relation de base entre la distance $d$, la vitesse $V$, et le temps $t$ :

$d = V \cdot t$

Nous pouvons réorganiser cette équation pour calculer le temps $t$ :

$t = \frac{d}{V}$

Appliquons cette formule à chaque étape de la course :

1. La pente (h = 2 m, d = 31 m, $V_f$ = $19,97 m/s)$ :

$t_{pente} = \frac{31 \, \text{m}}{19,97 \, \text{m/s}}$

2. Le looping ($r = 6 m$, $V_f$ = $25,20 m/s$, $V_i$ = $19,97 m/s$) :

Pour calculer le temps dans le looping en utilisant la vitesse angulaire $\omega$, nous pouvons utiliser la formule pour la vitesse angulaire :

$\omega = \frac{V}{r}$

$\omega_f = \frac{V_f}{r}$

$\omega_i = \frac{V_i}{r}$

Le temps $t_{looping}$ nécessaire pour traverser le looping est donné par :

$t_{looping} = \frac{\theta}{\omega}$

Où $\theta$ est l'angle total du looping (360 degrés ou $2\pi$ radians). Donc,

$\theta = 2\pi$

$\omega_f = \frac{25,20 \, \text{m/s}}{6 \, \text{m}}$

$\omega_i = \frac{19,97 \, \text{m/s}}{6 \, \text{m}}$

$t_{looping} = \frac{2\pi}{\omega_f - \omega_i}$

3. **Le ravin ($v_i$ = $25,20 m/s$, $vf = 21,40 m/s$, $h = 1 m, d = 9 m$) :**

$t_{ravin} = \frac{9 \, \text{m}}{25,20 \, \text{m/s}}$

4. **La fin de la piste ($V_i$ = $21,40 m/s$, $d = 10 m$) :**

$t_{fin\_de\_piste} = \frac{10 \, \text{m}}{21,40 \, \text{m/s}}$

Maintenant, calculons chaque temps :

1. $t_{pente} \approx \frac{31}{19,97} \approx 1,55$ secondes
2. Pour le temps dans le looping, nous devons calculer l'angle total $\theta$ :

$\omega_f = \frac{25,20}{6} \approx 4,20 \, \text{rad/s}$
$\omega_i = \frac{19,97}{6} \approx 3,33 \, \text{rad/s}$

$t_{looping} = \frac{2\pi}{4,20 - 3,33} \approx 2,76$ secondes

3. $t_{ravin} \approx \frac{9}{25,20} \approx 0,36$ secondes
4. $t_{fin\_de\_piste} \approx \frac{10}{21,68} \approx 0,47$ secondes

Donc, le temps total nécessaire pour la voiture pour parcourir l'ensemble de la course serait la somme de ces temps pour chaque étape. Ajoutons-les :

Temps total = $t_{pente} + t_{looping} + t_{ravin} + t_{fin\_de\_piste}$

Temps total $\approx 1,55 + 2,76 + 0,36 + 0,47 \approx 4,04$ secondes

Il faudrait environ 4,04 secondes à la voiture pour terminer la course complète.

```python
import math

# Calculations for each section
t_slope = 31 / 19.97
omega_f_looping = 25.20 / 6
omega_i_looping = 19.97 / 6
theta_looping = 2 * math.pi
omega_avg_looping = (omega_f_looping + omega_i_looping) / 2
t_looping = theta_looping / omega_avg_looping
t_ravin = 9 / 25.20
t_end_of_track = 10 / 21.40

# Total time
total_time = t_slope + t_looping + t_ravin + t_end_of_track
print("le temps totale est de :", total_time)
```

# <font color="pink">Le calcule de la vitesse d'une Voiture dans la Pente avec Frottement : </font>

Ces calcule explore la dynamique d'une voiture descendant une pente en tenant compte du frottement. Le but est de déterminer la vitesse de la voiture après avoir parcouru une certaine distance en utilisant les principes de base de la physique.

## <font color="pink">Equation du mouvement : </font>

La dynamique de la voiture est décrite en utilisant l'équation différentielle suivante pour la vitesse sur l'axe x (horizontal):

$\frac{dv_x}{dt} = g(\sin(\alpha) - u \cos(\alpha)) - \frac{1}{2}k \frac{v_x^2}{m} + 5.81$

- **Masse de la voiture (m):** 1540 kg
- **Coefficient de frottement (u):** 0.1
- **Accélération due à la gravité (g):** 9.8 m/s²
- **Coefficient de portance (k):** 0.8
- **Angle de la pente (alpha):** 0.063 radians
- **Delta de temps (dt):** 0.01 s, pour la résolution numérique
- **Temps final (tF):** 3.5 s, durée totale de la simulation

Cette équation prend en compte :
- La force gravitationnelle contribuant à l'accélération de la voiture le long de la pente.
- La force de frottement agissant contre le mouvement de la voiture.
- La résistance de l'air proportionnelle au carré de la vitesse de la voiture.

Donc la vitesse de la voiture après avoir parcouru une distance spécifiée est de $18,21 m/s$.

Le graphe ci dessus affiche la vitesse de la voiture lors de son passage dans la pente avec frottement : 

```python
import numpy as np
import matplotlib.pyplot as plt

# Paramètres
alpha_degrees = 3.7  # Angle de tir en degrés
mu = 0.1  # Coefficient de frottement
k = 0.5  # Constante de résistance de l'air
m = 1540  # Masse du projectile
v0 = 0  # Vitesse initiale en m/s
x0 = 0.0  # Position initiale en m
g = 9.81  # Accélération due à la gravité en m/s^2
am = 5.8  # Accélération externe en m/s^2

# Discrétisation du temps
dt = 0.01  # Pas de temps en secondes
num_steps = 340  # Nombre d'itérations

# Initialisation des listes pour stocker les valeurs calculées
x_values = [x0]
v_values = [v0]

# Calcul des valeurs en utilisant la méthode d'Euler
for i in range(num_steps):
    t = i * dt
    v = v_values[-1]
    x = x_values[-1]
    
    # Calcul de l'accélération
    a = g * (-mu * np.cos(np.radians(alpha_degrees)) + np.sin(np.radians(alpha_degrees))) + am - (k / m) * v**2
    
    # Mise à jour de la vitesse et de la position en utilisant la méthode d'Euler
    v += a * dt
    x += v * dt
    
    # Ajout des nouvelles valeurs aux listes
    v_values.append(v)
    x_values.append(x)

# Calcul du temps total et de la vitesse finale
temps_total = num_steps * dt
vitesse_finale = v_values[-1]

# Affichage du temps total et de la vitesse finale
print(f"Temps total pris : {temps_total} secondes.")
print(f"Vitesse finale du la voiture : {vitesse_finale:.2f} m/s.")

# Tracé de la vitesse en fonction de la distance
plt.plot(x_values, v_values, color="pink", label="Vitesse")
plt.xlabel('Distance (m)')
plt.ylabel('Vitesse (m/s)')
plt.title('Vitesse en fonction de la distance')
plt.legend()
plt.grid(True)
plt.show()
```

## <font color="pink">Le calcule de la Vitesse de la Voiture dans le Looping avec Frottement : </font>

Ces calcules examinent comment la vitesse d'une voiture évolue lorsqu'elle parcourt un looping, en prenant en compte les effets du frottement. Nous utilisons l'équation différentielle du mouvement pour modéliser la situation et résoudre numériquement pour la vitesse angulaire et la vitesse linéaire de la voiture à travers le looping.

## <font color="pink">Equation du mouvement : </font>

L'équation de mouvement pour la voiture dans le looping est dérivée en prenant en compte les forces agissant sur la voiture, y compris la gravité, le frottement, et d'autres forces spécifiques au mouvement circulaire. L'équation différentielle utilisée est la suivante :

$\frac{d^2\theta}{dt^2} = \frac{5.8 \cdot m + m \cdot g \cdot (-\sin(\theta) - \mu \cdot \cos(\theta)) - \left( \frac{d\theta}{dt} \right)^2 \cdot (\mu \cdot m \cdot r + k \cdot r^2)}{m \cdot r}$

où :
- $(\theta)$ est l'angle de rotation en radians,
- $(\frac{d\theta}{dt})$ est la vitesse angulaire de la voiture,
- $(m)$ est la masse de la voiture,
- $(g)$ est l'accélération due à la gravité,
- $(\mu)$ est le coefficient de frottement,
- $(r)$ est le rayon du looping,
- $(k)$ est un paramètre supplémentaire reflétant d'autres forces ou caractéristiques,
- $(5.8 \cdot m)$ représente une force constante agissant sur la voiture (peut-être une simplification pour inclure des effets comme la propulsion ou d'autres forces appliquées).

Cette équation prend en compte à la fois la composante gravitationnelle agissant sur la voiture en mouvement circulaire, les effets du frottement entre la voiture et la surface du looping, et d'autres forces potentielles exprimées par le terme en $(k)$. La résolution de cette équation différentielle nous permet de calculer la vitesse angulaire $(\frac{d\theta}{dt})$ à tout moment $(t)$, que nous pouvons ensuite utiliser pour déterminer la vitesse linéaire de la voiture en multipliant par le rayon $(r)$ du looping.

Donc la vitesse initiale avant d'entrée dans le looping est de $( 18,31 \, \text{m/s}^-1)$, et la vitesse finale apres avoir sortie du looping est de $(20,51 \, \text{m/s}^-1)$

Ce graphe ci dessus represente la vitesse initiale et finale de notre voiture avec frottement : 


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Paramètres
donnees = {
    "m": 1540,       # Masse de la voiture en kg
    "g": 9.81,       # Accélération gravitationnelle en m/s^2
    "mu": 0.1,       # Coefficient de frottement
    "r": 6,          # Rayon du looping en mètres
    "k": 0.5,        # Paramètre supplémentaire
    "t0": 0,         # Temps initial
    "tmax": 2.3,     # Temps maximal
    "pas": 1000      # Nombre de points pour la discrétisation temporelle
}

v_plat_af1 = 18.59  # Vitesse initiale de la voiture en m/s

# Equation de mouvement à résoudre avec 'odeint'
def equation(y, t):
    calcul = (donnees["m"] * donnees["g"] * (-np.sin(y[0]) - donnees["mu"] * np.cos(y[0])) - (y[1] ** 2) * (
                donnees["mu"] * donnees["m"] * donnees["r"] + donnees["k"] * donnees["r"] ** 2)+5.8*1540) / (
                         donnees["m"] * donnees["r"])
    return [y[1], calcul]

# Vecteur temps
t = np.linspace(donnees["t0"], donnees["tmax"], donnees["pas"])

# Calcul vitesse angulaire (teta prime) en rad/s^-1
omega = v_plat_af1 / donnees["r"]

# Résolution
y = odeint(equation, [0, omega], t)

# Savoir quand est-ce qu'on dépasse 2pi
i = 0
while i < len(y) and y[i][0] < 2 * np.pi:
    i += 1

temps_complet = t[i - 1]  # Temps pris pour atteindre la fin du looping
v_looping_af = y[i - 1][1] * donnees["r"]  # Vitesse à la fin du looping

print(f"Temps pour compléter le looping : {temps_complet:.2f} s")
print(f"Vitesse initiale : {v_plat_af1:.2f} m/s | Vitesse finale sortie de looping : {v_looping_af:.2f} m/s")

# Ajustement du vecteur temps pour affichage complet jusqu'à 'i'
t_ajuste = np.linspace(donnees["t0"], temps_complet, i)

# Tracer des résultats
plt.plot(t_ajuste, y[0:i, 1] * donnees["r"], color="pink", label="Vitesse avec frottement")

# Titre et noms des axes
plt.title("Vitesse de la voiture avec frottements en fonction du temps")
plt.xlabel("Temps (s)")
plt.ylabel("Vitesse (m/s)")

plt.grid()
plt.legend()
plt.show()
```

## <font color="pink">Calcule de la vitesse lors de la Trajectoire au Ravin avec Frottement : </font>

Ces calcules étudient la vitesse lors de la trajectoire de notre voiture traversant un ravin, en prenant en compte les forces de frottement et de portance. En utilisant une approche numérique, nous calculons les vitesses horizontale et verticale ainsi que la position de la voiture au fil du temps.

## <font color="pink">Equation du mouvement : </font>

Les équations différentielles suivantes décrivent les forces agissant sur la voiture dans les directions horizontale (x) et verticale (y) :

- **Sur l'axe x :**
$\frac{dv_x}{dt} = -\frac{1}{2m} \cdot (\ cx \cdot \sqrt{v_x^2 + v_y^2} \cdot v_x + cy \cdot \sqrt{v_x^2 + v_y^2} \cdot v_x$

- **Sur l'axe y :**
$\frac{dv_y}{dt} = -g + \frac{1}{2m} \cdot (\ cx \cdot \sqrt{v_x^2 + v_y^2} \cdot v_y + cy \cdot \sqrt{v_x^2 + v_y^2} \cdot v_y )$

Où : 

- **Masse de la voiture (m):** 1540 kg
- **Coefficient de traînée (cx):** 0.8
- **Coefficient de portance (cy):** 2.4
- **Accélération due à la gravité (g):** 9.81 m/s²
- **Temps de simulation (tF):** 1 seconde
- **Pas de temps (dt):** 0.01 seconde
- **Vitesse initiale en sortie du looping avec frottements (Vsl_f):** 20.51 m/s

Ces équations tiennent compte de la résistance de l'air (modélisée par le coefficient de traînée cx) et de la force de portance (modélisée par le coefficient cy), en plus de la gravité. La résolution de ces équations nous permet de simuler la trajectoire de la voiture à travers le ravin.

Finalement, La vitesse de la voiture par rapport au temps au ravin est de $(20,31 \, \text{m/s})$ et le temps utilisé est de (0,45 secondes)

Les graphes ci dessus representent La trajectoire de notre voiture au ravin et la vitesse dez notre voiture lors du saut au ravin et sa position par rapport a l'axe des abscisses : 


```python
import numpy as np
import matplotlib.pyplot as plt

m = 1540  # masse de la voiture
cx = 0.8  # coefficient de traînée
g = 9.81
cy = 2.4  # coefficient de portance
dt = 0.01  # delta t
tF = 1  # Temps final
Vsl_f = 20.51  # la vitesse en sortie du looping avec frottements
Vx0 = Vsl_f
vy0 = 0  # vitesse initiale sur y
vx = [Vx0]  # tableau des Vx
vy = [vy0]  # tableau des Vy
t = np.arange(0, tF, dt)  # axe du temps

# Déclaration de la fonction différentielle renvoyant à la vitesse sur l'axe x
def dvxdt(vx, vy):
    return (-(1 / 2 * cx * np.sqrt(vx ** 2 + vy ** 2) * vx + (1 / 2) * cy * np.sqrt(vx ** 2 + vy ** 2) * vx)) / m

# Déclaration de la fonction différentielle renvoyant à la vitesse sur l'axe y
def dvydt(vx, vy):
    return (-g + (1 / 2 * cx * np.sqrt(vx ** 2 + vy ** 2) * vy + 1 / 2 * cy * np.sqrt(vx ** 2 + vy ** 2) * vy) / m)

# Resolution avec la methode d'Euler
for i in range(0, int(tF / dt) - 1):
    vx.append(vx[i] + dvxdt(vx[i], vy[i]) * dt)
    vy.append(vy[i] + dvydt(vx[i], vy[i]) * dt)

# Initialisation des variables pour la trajectoire
x0 = 0
y0 = 0

# Génération des tableaux pour la trajectoire
x_f = [x0]
y_f = [y0]

# Resolution avec méthode numérique (Euler)
for i in range(0, int(tF / dt) - 1):
    x_f.append(x_f[i] + vx[i] * dt)
    y_f.append(y_f[i] + vy[i] * dt)

# Recherche de l'indice es
indice_x_9 = np.argmax(np.array(x_f) >= 9)
indice_t_9 = np.argmax(np.array(x_f) >= 9)
indice_y_minus_1 = np.argmax(np.array(y_f) <= -1)

# Affichage de la valeur correspondante
x_minus_1 = x_f[indice_y_minus_1]
y_9 = y_f[indice_x_9]
t_9 = t[indice_x_9]
v_9 = vx[indice_t_9]
# Affichage de la valeur de x lorsque y = -1 mètre et la valeur de y lorsque x = 9 mètres
print("Le temp total est :", t_9,"S")
print("La vitesse final aprés le saut:", v_9,"M/S")

# Affichage de la trajectoire dans le ravin avec frottements
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# Trajectoire
axs[0].plot(x_f, y_f, 'pink')
axs[0].set_xlabel("x")
axs[0].set_ylabel("y")
axs[0].set_title("Trajectoire dans le ravin avec frottements")
axs[0].grid()

# Vitesse par rapport au temps
axs[1].plot(t, vx, 'pink')
axs[1].set_xlabel("Temps (s)")
axs[1].set_ylabel("Vitesse (m/s)")
axs[1].set_title("Vitesse par rapport au temps")
axs[1].grid()

# Vitesse par rapport au temps (graphique 2)
axs[2].plot(t, x_f, 'pink')
axs[2].set_xlabel("Temps (s)")
axs[2].set_ylabel("Position x (m)")
axs[2].set_title("Position x par rapport au temps")
axs[2].grid()

plt.show()
```

## <font color="pink">Calcule de la Vitesse de la voiture lors de l'arrivé a la fin de la piste avec Frottements : 
  </font>

Ces calcules montrent la vitesse et la trajectoire d'une voiture sur une piste plate en prenant en compte l'effet des frottements. L'objectif est de déterminer comment la vitesse de la voiture évolue avec le temps et à quelle distance elle se trouve après un certain temps.

## <font color="pink"> Equations du mouvement :  
  </font>

  $\frac{dv_x}{dt} = \frac{1}{2} k \frac{v_x^2}{m} + \text{Constante}$

  Où :

  - $(v_x)$ est la vitesse de la voiture sur l'axe x,
- $(m)$ est la masse de la voiture,
- $(k)$ est le coefficient de frottement,
- La constante représente une force externe agissant sur la voiture (ici, 5.2).

Cette équation décrit comment la vitesse de la voiture évolue en raison des forces de frottement et d'une force constante appliquée.

Donc la vitesse utilisé lors de l'arrivé a la fin de la piste est de $(22,81 \, \text{m/s})$ et le temps utilisé est de (0,47 secondes)

le ghraphe ci dessus representent La vitesse a l'arrivée de la fin de la piste :

```python
import numpy as np
import matplotlib.pyplot as plt

#Paramètres
alpha = np.radians(3.7)  # Angle de tir en radians
mu = 0.1  # Coefficient de frottement
k = 0.5  # Constante de résistance de l'air
m = 1540  # Masse du projectile
v0 = 20.4  # Vitesse initiale en m/s
x0 = 0  # Position initiale en m
g = 9.81  # Accélération due à la gravité en m/s^2
am = 5.8  # Accélération externe en m/s^2

#Discrétisation du temps
dt = 0.01  # Pas de temps en secondes
num_steps = 200  # Nombre d'itérations

#Initialisation des listes pour stocker les valeurs calculées
x_values = [x0]
v_values = [v0]

#Calcul des valeurs en utilisant la méthode d'Euler
for i in range(num_steps):
    t = i * dt
    v = v_values[-1]
    x = x_values[-1]

    # Calcul de la vitesse et de la position
    v_x = (am - mu * g - k / m * v) * t + v0
    x_t = ((am - mu * g - k / m * v) / 2) * t*2 + v0*t + x0

    # Ajout des nouvelles valeurs aux listes
    v_values.append(v_x)
    x_values.append(x_t)

    # Arrêt lorsque la distance atteint 10 mètres
    if x_t >= 10:
        break

#Tracé de la vitesse en fonction de la distance avec un libellé
plt.plot(x_values, v_values, color="pink", label='Vitesse vs Distance')
plt.xlabel('Distance (m)')
plt.ylabel('Vitesse (m/s)')
plt.title('Vitesse en fonction de la distance')
plt.grid(True)
plt.legend()
plt.show()

#Calcul du temps et de la vitesse pour une distance cible
x_target = 10.0
t_target = t  # Utilisation du dernier temps calculé
v_target = v_x  # Utilisation de la dernière vitesse calculée

print(f"Temps pour atteindre {x_target} m : {t_target:.2f} s.")
print(f"Vitesse à cette distance : {v_target:.2f} m/s.")
```
